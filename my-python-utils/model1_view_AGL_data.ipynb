```python
d1 = feature[3121,0,:,:]
d2 = feature[4152,5,:,:]
```

```python
Qp = np.percentile(d1, 99.9)
Qmin = np.amin(d1)
Qmax = np.amax(d1)
print(Qmin, Qmax, Qp)

Wp = np.percentile(d2,99.9)
Wmin = np.amin(d2)
Wmax = np.amax(d2)
print(Wmin, Wmax, Wp)

cmap = plt.cm.Spectral_r

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16,6))

f = ax1.imshow(d1, cmap=cmap)
ax1.set_title("Normalized QRAIN")
f.set_clim(vmin=Qmin, vmax=Qmax)
fig.colorbar(f, ax=ax1, orientation='vertical', label='')

l = ax2.imshow(d2, cmap=cmap)
ax2.set_title("Normalized W")
l.set_clim(vmin=Wmin, vmax=Wmax)
fig.colorbar(l, ax=ax2)
```
